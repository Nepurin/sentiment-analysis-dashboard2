```python
from transformers import pipeline
import plotly.express as px
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sentiment_model = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment"
)
```

```python
text = "Saya suka produk ini, kualitasnya sangat bagus!"
result = sentiment_model(text)
result
```

```python
comments = [
    "Produk ini jelek sekali, tidak sesuai harapan",
    "Pengiriman cepat dan kualitas oke",
    "Harga terlalu mahal dibanding toko lain",
    "Sangat puas dengan pelayanan customer service",
    "Biasa saja, tidak terlalu istimewa",
    "Barang cacat, tidak akan beli lagi",
    "Super recommended! Kualitas premium dengan harga bagus"
]

results = sentiment_model(comments)

# Gabungkan ke DataFrame
df = pd.DataFrame(results)
df['comment'] = comments
df
```

```python
def convert_label(label):
    if "1 star" in label or "2 star" in label:
        return "Negatif"
    elif "3 star" in label:
        return "Netral"
    else:
        return "Positif"

df['sentiment'] = df['label'].apply(convert_label)
df
```

```python
fig = px.pie(df, names='sentiment', title="Distribusi Sentimen Komentar")
fig.show()
```

```python
all_text = " ".join(comments)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
